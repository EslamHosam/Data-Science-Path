## Multiple plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read your data
unrate = pd.read_csv("unrate.csv")
unrate.head()
```

```python
# some information about the data
unrate.info()
```

```python
# convert the DATE column to datetime 
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
```

```python
#prepare x , y for your plot
x = unrate.loc[:, "DATE"].iloc[:12]
y = unrate.loc[:, "VALUE"].iloc[:12]
```

```python
#DISPLAY YOUR PLOT
plt.plot(x, y)
plt.xlabel("DATE")
plt.xticks(rotation = 90)
plt.ylabel("VALUE")
plt.title("Unemployment Rate in one year")
plt.show()
```

## matplotlib classes

```python
# create the first subplot with the plt.figure object
# first create a figure
fig = plt.figure()
#then create a subplot 
ax_1 = fig.add_subplot(2,1,1)
#add another subplot
ax_2 = fig.add_subplot(2,1,2)

plt.show()
```

```python
# try to nderstand the grid postioning
# create more plots
fig = plt.figure()
ax_1 = fig.add_subplot(2,2,1)
ax_2 = fig.add_subplot(2,2,2)
ax_3 = fig.add_subplot(2,2,3)
ax_4 = fig.add_subplot(2,2,4)

plt.show()
```

- When the first subplot is created, matplotlib knows to create a grid with 2 rows and 2 columns. As we add each subplot, we specify the plot number we want returned and the corresponding Axes object is created and returned. In matplotlib, the plot number starts at the top left position in the grid (left-most plot on the top row), moves through the remaining positions in that row, then jumps to the left-most plot in the second row, and so forth.

- If we created a grid of 4 subplots but don't create a subplot for each position in the grid, areas without axes are left blank:



```python
fig = plt.figure()
ax_1 = fig.add_subplot(2,2,1)
ax_3 = fig.add_subplot(2,2,3)
ax_4 = fig.add_subplot(2,2,4)
plt.show()
```

## add data to subplots

```python
# create 2 subplots in a 2 row by 1 column layout:
fig = plt.figure()
#create subplots
ax_1 = fig.add_subplot(2,1,1)
ax_2 = fig.add_subplot(2,1,2)
#add the data to your plot using Axis.plot()
#plotting the data for first year 1948
ax_1.plot(x,y)
#plot the data for second year
ax_2.plot(unrate["DATE"].iloc[12:24] , unrate["VALUE"].iloc[12:24])
plt.show()
```

## formating and spacing and customizing the apperance

```python
# specify the dimensions of the plots using figsize argument
fig = plt.figure(figsize = (12,5))
ax_1 = fig.add_subplot(2,1,1)
ax_2 = fig.add_subplot(2,1,2)
ax_1.plot(x,y)
ax_2.plot(unrate["DATE"].iloc[12:24] , unrate["VALUE"].iloc[12:24])
plt.show()
```

```python
# we can compare between more years not just 2 or 3 years in 3 plots
#create multiple plots to compare 5 years
#create the figure
fig = plt.figure(figsize = (12,12))

#use for loop to create the subplots
for i in range(5):
    ax = fig.add_subplot(5,1, i+1)
    start = i * 12 
    end = (i +1) * 12
    ax.plot(unrate["DATE"].iloc[start : end] , unrate["VALUE"].iloc[start : end])
    
plt.show()
```

```python
# customize your ploting for months in the x axis instead of years
# add new column to your data
unrate["MONTH"] = unrate["DATE"].dt.month
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'])
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'])
```
