# bar plot in matplotlib and scatter plots

```python
dataset = pd.read_csv("fandango_scores.csv")
reviews = dataset[["FILM" , "RT_user_norm" , "Metacritic_user_nom" , "IMDB_norm" , "Fandango_Ratingvalue" , "Fandango_Stars"]]
reviews.head()
```

```python
# bar plots 
fig = plt.figure(figsize = (7,5))

names = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm','Fandango_Ratingvalue', 'Fandango_Stars']
values = reviews.iloc[0].values

ax = fig.add_subplot(111)
ax.bar(names, values[1:] , 0.5)

ax.set_xticks(np.arange(5))
ax.set_xticklabels(names, rotation = 45) 

plt.show()
```

```python
fig,ax = plt.subplots(sharex=True)
x = reviews.iloc[:5]["FILM"]
y1 = reviews.iloc[:5]["Metacritic_user_nom"]
y2 = reviews.iloc[:5]["IMDB_norm"]


ax.bar(x,y1, 0.3, align="edge", )
ax.bar(x,y2, 0.3, align="center")

```

```python
reviews.iloc[:5].plot(x= "FILM" , y = ["RT_user_norm", "IMDB_norm"] , kind = "bar" )
```

```python
# scatter plot 
x = reviews.iloc[:5]["IMDB_norm"]
y = reviews.iloc[:5]["RT_user_norm"]

fig, ax = plt.subplots()

ax.scatter(x,y, label = "A")
ax.scatter(x, y+2, label = "B")
ax.set_title("Scatter plot")
ax.set_xlabel("IMDB")
ax.set_ylabel("RT_user")
ax.legend(loc = "lower right")
plt.show()
```

```python
num_cols = ["RT_user_norm" , "Metacritic_user_nom" , "IMDB_norm" , 
            "Fandango_Ratingvalue" , "Fandango_Stars"]
bar_positions = np.arange(5)

bar_heights = reviews[num_cols].iloc[0].values


fig, ax = plt.subplots(figsize = (7,5))

ax.bar(bar_positions, bar_heights, 0.5)

ax.set_xticks(np.arange(5))
ax.set_xticklabels(num_cols, rotation = 90)

plt.show()
```

# Guide project visualize earnings based on college majors 

```python
# importing libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from pandas.plotting import scatter_matrix
import seaborn as sns
```

```python
# read and discover the dataset 
dataset = pd.read_csv("recent-grads.csv")
dataset.head()
```

```python
# grt your information 
dataset.info()
```

```python
# check for null values 
dataset.isnull().sum()
```

```python
# drop all missing values because they are very less 
majors = dataset.dropna(axis= 0)
majors.head()
```

```python
print("original number of rows in our dataset {}".format(dataset.shape[0]))
print("number of rows after removing missing values {}".format(majors.shape[0]))
```

```python
# describe the data or get some information 
majors.describe().T
```

## pandas scatter plots

```python
majors.plot(x= "Men", y= "Women" , kind = "scatter", title = "men vs women",
           figsize = (7,5))
```

```python
fig = plt.figure(figsize = (7,5))
ax = fig.add_subplot(111)

ax.scatter(majors["Men"], majors["Women"], c= "red")
ax.set_title("compare women and men correlation")
ax.set_xlabel("men")
ax.set_ylabel("women")

plt.show()
```

```python
# check the coelation between the men and the median salary 
majors.plot(x="Men", y= "Median", kind = "scatter", figsize = (7,5), c= "green",
           title = "men and salary" , rot = 90 , fontsize = 15, colormap="magma", colorbar = False, 
           marker = 5, )
```

```python
# scatter plot shows the distribution of the data over a column and the correlation between columns
majors.plot(x="Sample_size", y="Employed", kind = "scatter", figsize= (10,5), title= "sample and employed correlation",
          fontsize = 12)
```

```python
majors.plot(x="Men", y="Rank", kind="scatter", figsize = (8,6), fontsize=13, colormap="magma", colorbar=False)
```

```python
fig,ax = plt.subplots(figsize=(7,7))

ax.scatter(majors["Men"], majors["Sample_size"])
ax.set_xlabel("Men")
ax.set_ylabel("sample size")
```

```python
# save the Axis object in a variable and use it normally with matplotlib 
ax = majors.plot(x="Men" , y= "Median" , kind = "scatter", figsize = (8,5))

ax.set_xlabel("men are here")
ax.set_ylabel("m e d i a n")
ax.set_title("men and median relationship")

plt.show()
```

```python
# do students in more popular majors make more money 
ax = majors.plot(x="Total" , y="Median", kind = "scatter", figsize = (8,5), cmap="magma", colorbar= False)
ax.set_title("median salary for majors")
```

```python
d1 = majors[majors["Women"] > 5000]
d1.plot(x="Women", y="Median" , kind = "scatter", figsize =(12,5))
```

```python
grouped = majors.groupby("Major_category").mean()
grouped
```

```python
ax = grouped["Median"].plot(kind="barh", figsize = (15,8), color="orange", fontsize=15)

ax.set_title("median salary for major categories")
ax.set_xlabel("salary")
```
