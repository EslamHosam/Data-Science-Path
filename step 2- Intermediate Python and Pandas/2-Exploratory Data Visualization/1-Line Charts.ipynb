```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
unrate = pd.read_csv("unrate.csv")
unrate.head()
```

```python
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
```

```python
unrate["DATE"].dtypes
```

```python
x = unrate.iloc[:12, 0]
y = unrate.iloc[:12, 1]
```

```python
plt.plot(x,y)
plt.xlabel("date")
plt.ylabel("value")
plt.title("unemployment rate")
# plt.xticks(rotation = 90)
plt.show()
```

## matplotlib classes

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(x,y)
x2 = unrate["DATE"].iloc[12:24]
y2 = unrate["VALUE"].iloc[12:24]

ax2.plot(x2,y2)
plt.show()
```

## grid positioning

```python
fig = plt.figure()
a = fig.add_subplot(2,1,1)
b = fig.add_subplot(2,1,2)
a.plot(x,y)
b.plot(x,y)
```

```python
fig = plt.figure()

a = fig.add_subplot(2,1,1)
a.plot([1,2,3] , [3,5,6])
plt.show()
```

### formating and spacing or Axis.plot()

```python
fig = plt.figure(figsize = (12,5))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(unrate[:11]["DATE"] , unrate[:11]["VALUE"])
ax2.plot(unrate[12:24]["DATE"] , unrate[12:24]["VALUE"])

ax1.set_title("monthly unemployment rate")
ax2.set_title("monthly unemplyment rate 1949")


plt.show()
```

## using loops to plot 

```python
fig = plt.figure(figsize = (12,5))

for i in range(2):
    ax = fig.add_subplot(2,1,i+1)
    start_index = i*12
    end_endex = (i+1) * 12
    subset = unrate[start_index : end_endex]
    ax.plot(subset["DATE"] , subset["VALUE"])
    
plt.show()
```

```python
fig = plt.figure(figsize = (12,12))

for i in range(5):
    ax = fig.add_subplot(5,1, i+1)
    start_index = i* 12
    end_index = (i+1) * 12
    z = unrate[start_index : end_index]
    ax.plot(z["DATE"] , z["VALUE"])
    
plt.show()
```
