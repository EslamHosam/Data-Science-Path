```python
# visualizing the STEM fields 
major_cats = ["Biology" , "Computer Science" , "Engineering" , "Math and Statistics"]
fig = plt.figure(figsize = (12,12))

for i in range(4) : 
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees["Year"] , women_degrees[major_cats[i]] , c= "blue", label="women")
    ax.plot(women_degrees["Year"] , 100 - women_degrees[major_cats[i]], c="green" , label= "men")
    #title 
    ax.set_title(major_cats[i])
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    
    
    # hide xticks and spines
    ax.tick_params(top =0, bottom = 0, right= 0, left= 0)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
#     ax.spines["top"].set_visible(False)
#     ax.spines["right"].set_visible(False)
#     ax.spines["left"].set_visible(False)
#     ax.spines["bottom"].set_visible(False)
    
# display the legend out of for loop 
ax.legend(loc = "best")

plt.show()
```

```python
# visualizing the gender gabs over the years using line charts 
fig = plt.figure(figsize=(20,5))

green = (120/255 , 205/255, 50/255)
orange = (1 , 108/255, 14/255)

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

for i in range(6):
    ax = fig.add_subplot(1,6, i+1)
    x = women_degrees["Year"]
    y1 = women_degrees[stem_cats[i]]
    y2 = 100 - y1
    
    # plot easily and clearly 
    ax.plot(x, y1, c= green, linewidth= 2.3)
    ax.plot(x, y2, c=orange, linewidth= 2.3)
    
    # set the axis limits 
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    # improving the plot apperance 
    # removing all ticks 
    ax.tick_params(left=0, top= 0, right=0, bottom=0)
    
    # removing all the spines 
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    # adding text instead of legend in the plots 
    if i == 0 :
        ax.text(x= 2005, y= 74, s="Men")
        ax.text(2005,25,"Women")
        
    elif i == 2 :
        ax.text(2005,67, "Women")
        ax.text(2005,30,"men")
        
    elif i ==5:
        ax.text(2010,59, "men")
        ax.text(2010,40,"women")
        
plt.show()
```

```python
# put the groups together 
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance',
                 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(20,20))

green = (120/255 , 205/255, 50/255)
orange = (1 , 108/255, 14/255)

# plot the first list stem_cats 

for i in range(6):
    ax = fig.add_subplot(6,3, (i*3) +1)
    # plotting 
    x = women_degrees["Year"]
    y1 = women_degrees[stem_cats[i]]
    y2 = 100 -y1
    
    ax.plot(x,y1, c=green , lw=2, ls=None)
    ax.plot(x,y2, c= orange, lw=2,ls= None)
    
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    
    # improving plotting 
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["right"].set_visible(False)
    
    ax.tick_params(left = False, bottom = False)
    if i != 5:
        ax.tick_params(labelbottom = False)
        
    # add horizntal line 
    ax.axhline(y = 50 , alpha = 0.2)
    
    # annotate the plot with text 
    if i == 0 :
        ax.text(2005 , 84, "women")
        ax.text(2005, 11, "men")
    # adding title of each plot 
    ax.set_title(stem_cats[i])
#     ax.set_yticks([0,100])
    
# for the second list of categories lib_arts_cats we need 5 plots in the second column 
for i in range(5):
    ax = fig.add_subplot(6,3 ,(i*3)+2)
    # x axis is the same years 
    y1 = women_degrees[lib_arts_cats[i]]
    y2 = 100 - y1
    ax.plot(x,y1, lw=2 , c= green , ls= None , label = "women")
    ax.plot(x,y2 , lw= 2, ls = None , c= orange, label = "men")
    
    # improving the plot 
    ax.set_title(lib_arts_cats[i])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    
    # removing  ticks and spines 
    ax.tick_params(left= False, top = False, bottom = False, right =False , labelbottom = False )
    for k,spine in ax.spines.items():
        spine.set_visible(False)
    
    # adding bottom label to last plot 
    if i == 4 :
        ax.tick_params(labelbottom = True)
        
    ax.set_yticks([0,100])
    
    # annotating the lines with text 
    if i == 0 :
        ax.text(2005 , 85, "women", c = green)
        ax.text(2005, 20, "men" , c= orange)
    elif i == 4:
        ax.text(1975,65,"men" , c = orange)
        ax.text(1975, 25, "women" , c= green)
    # adding horizental line 
    ax.axhline(y=50 , alpha = 0.2)
    
    
#  plot for the last list other_cats we need 6 plot s also but in 3rd column 
for i in range(6):
    ax = fig.add_subplot(6,3, (3*i)+3)
    y1 = women_degrees[other_cats[i]]
    y2 = 100 - y1
    
    ax.plot(x,y1, c=green, lw=2, ls= None, label = "women")
    ax.plot(x,y2, c=orange, lw= 2, ls=None, label = "men")
    
    # improving the plot 
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    # titles 
    ax.set_title(other_cats[i])
    
    #removing spins and ticks
    ax.tick_params(left= False,right=False,top=False,bottom = False , labelbottom = False)
    for k,spine in ax.spines.items():
        spine.set_visible(False)
        
#     adding annotation to the lines 
    if i == 0 :
        ax.text(2005,75,"women" , c= green)
        ax.text(2005,22,"men" , c= orange)
    elif i == 5 :
        ax.text(2005,60,"men")
        ax.text(2005,20,"women")
        ax.tick_params(labelbottom = True)
        
    ax.axhline(y= 50, alpha = 0.2)
    
    
# saving the graph as png por jpg image 
plt.savefig(fname = "gender gap in college majors.jpg" ,dpi = 70)
plt.show()
```
