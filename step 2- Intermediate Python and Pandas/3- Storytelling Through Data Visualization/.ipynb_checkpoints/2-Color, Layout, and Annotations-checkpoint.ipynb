# Color, Layout, and Annotations
- In this mission, we'll focus more directly on customizing colors, line widths, layout, and annotations to improve the ability for a viewer to extract insights from the charts. We'll continue to use the same data set containing the percentage of bachelor's degrees granted to women from 1970 to 2012

```python
# importin tools and data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#r readind data
women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degrees.head(6)
```

```python
# put the major categories in one list
major_cat = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]

# define a figure and Axes
fig = plt.figure(figsize = (12,12))

#use for loop and plot graphs for gender gap for major categories
for i in range(0,4):
    ax = fig.add_subplot(2,2, i+1)
    ax.plot(women_degrees["Year"], women_degrees[major_cat[i]], label = "women")
    ax.plot(women_degrees["Year"], 100- women_degrees[major_cat[i]], label = "men")
    
    #set xlim and ylim
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    
    # hide all ticks
    ax.tick_params(left = False, right =False, bottom = False, top = False)
    # hide the spines
    for key, value in ax.spines.items():
        value.set_visible(False)
        
# adding legend to right most plot
ax.legend()
plt.show()
```

##  Setting Line Color Using RGB and Line Width

```python
# put the major categories in one list
major_cat = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]

# define a figure and Axes
fig = plt.figure(figsize = (12,12))

#use for loop and plot graphs for gender gap for major categories
for i in range(0,4):
    ax = fig.add_subplot(2,2, i+1)
    
    #set the colors
    ax.plot(women_degrees["Year"], women_degrees[major_cat[i]], label = "women", c = (0/255, 107/255, 164/255), 
           linewidth = 2.5)
    ax.plot(women_degrees["Year"], 100- women_degrees[major_cat[i]], label = "men", c = (255/255, 128/255, 14/255), 
           linewidth =2.5)
    #set x , ylimits
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    # hide ticks
    ax.tick_params(left = False, right =False, bottom = False, top = False)
    
    # hide spines
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
# adding legend to last plot
ax.legend(loc ="upper right")
plt.show()
```

## Improve the Layout and Ordering
- use layout of a single row with multiple columns
- order the plots in decreasing order of initial gender gap
- We've populated the list stem_cats with the six STEM degree categories, ordering them by decreasing ending gender gap. In the next step, we'll explore how we can replace the legend, which is currently overlapping with the rightmost line chart.

```python
# put all STEM categories
stem_cat = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

# define a figure and Axes
#Change the width of the figure to a width of 18 inches and a height of 3 inches.
fig = plt.figure(figsize = (18,3))

#use for loop and plot graphs for gender gap for major categories
# In the for loop, change the range to (0,6) instead of (0,4).
#Change the subplot layout from 2 rows by 2 columns to 1 row by 6 columns.
for i in range(0,6):
    ax = fig.add_subplot(1,6, i+1)
    
    #set the colors
    ax.plot(women_degrees["Year"], women_degrees[stem_cat[i]], label = "women", c = (0/255, 107/255, 164/255), 
           linewidth = 2.5)
    
    ax.plot(women_degrees["Year"], 100- women_degrees[stem_cat[i]], label = "men", c = (255/255, 128/255, 14/255), 
           linewidth =2.5)
    
    # set xlim and ylim
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    # hide ticks
    ax.tick_params(left = False, right =False, bottom = False, top = False)
    # hide spines
    for key, spine in ax.spines.items():
        spine.set_visible(False)
ax.legend(loc ="upper right")
plt.show()
```

### Replacing the Legend With Annotations
- To add text annotations to a matplotlib plot, we use the Axes.text() method. This method has a few required parameters:

- x: x-axis coordinate (as a float)
- y: y-axis coordinate (as a float)
- s: the text we want in the annotation (as a string value)
- 
- Add the following text annotations in the leftmost chart:
- The string "Men" at the x-axis coordinate of 2005 and the y-axis coordinate of 87.
- The string "Women" at the x-axis coordinate of 2002 and the y-axis coordinate of 8.

- Add the following text annotations in the rightmost chart:
- The string "Men" at the x-axis coordinate of 2005 and the y-axis coordinate of 62.
- The string "Women" at the x-axis coordinate of 2001 and the y-axis coordinate of 35.

- (tip) use if statement to add text if (i = 0 or 5)

```python
# put all STEM categories
stem_cat = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

# define a figure and Axes
#Change the width of the figure to a width of 18 inches and a height of 3 inches.
fig = plt.figure(figsize = (18,3))

for i in range(0,6):
    ax = fig.add_subplot(1,6, i+1)
    
    #set the colors
    ax.plot(women_degrees["Year"], women_degrees[stem_cat[i]], label = "women", c = (0/255, 107/255, 164/255), 
           linewidth = 2.5)
    
    ax.plot(women_degrees["Year"], 100- women_degrees[stem_cat[i]], label = "men", c = (255/255, 128/255, 14/255), 
           linewidth =2.5)
    # add annotation
    if i == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    
    if i == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")
    
    # set xlim and ylim
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    # set title for each graph
    ax.set_title(stem_cat[i])
    
    # hide ticks
    ax.tick_params(left = False, right =False, bottom = False, top = False)
    # hide spines
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
# removing the legend 
# ax.legend(loc ="upper right")
plt.show()
```
